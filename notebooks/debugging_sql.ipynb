```python
%%sql SELECT name AS "Athlete's name", COUNT(*)
    FROM medals
    GROUP BY name
    HAVING COUNT(*) > 0
    ORDER BY count DESC
```

```python
%sql SELECT * FROM medals WHERE country = 'Serbia'
```

```python
%%sql SELECT country,
    COUNT(*)
    FROM medals
    GROUP BY country
    ORDER BY count DESC;
```

```python
%%sql 
SELECT
    COUNT(*), country
    FROM medals
    GROUP BY country
    ORDER BY count DESC
```
