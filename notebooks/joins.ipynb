```python
%%sql SELECT CONCATa.name, b.date_of_birth, b.name
FROM athletes as a 
join athletes as b
ON DATE_PART('day', a.date_of_birth) = DATE_PART('day', b.date_of_birth)
AND DATE_PART('month', a.date_of_birth) = DATE_PART('month', b.date_of_birth)
ORDER BY "date_of_birth" DESC 
STRING_AGG(name)
LIMIT 10
```

```python
%%sql
SELECT  a.name, a.date_of_birth, b.date_of_birth, b.name
FROM athletes as a
join athletes as b on DATE_PART('day', a.date_of_birth) = DATE_PART('day', b.date_of_birth) 
AND DATE_PART('month', a.date_of_birth) = DATE_PART('month', b.date_of_birth)
GROUP BY a.date_of_birth, b.date_of_birth, a.name, b.name
HAVING a.name != b.name
LIMIT 20
```
